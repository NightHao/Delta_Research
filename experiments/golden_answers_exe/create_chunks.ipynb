```python
import pickle
with open('/home/yeskey525/Research_CODE/experiments/golden_answers/ISO_15118-3_chunk.pkl', 'rb') as f:
        content_ISO3 = pickle.load(f)
from anytree import Node, RenderTree
from anytree.exporter import DotExporter

# 篩選並構建樹狀結構
def build_tree(content):
    root = Node("Root")  # 根節點
    nodes = {}  # 存放已建立的node

    for item in content:
        current_parent = root
        for level in range(1, 8):  # iterate section_level_1 到 section_level_7
            key = f"section_level_{level}"
            value = item.get(key, "").strip()
            if value:  # 如果當前level有值
                if value not in nodes:  # 如果该node尚未建立
                    section_id = " > ".join(
                        [item.get(f"section_level_{lvl}", "").strip() for lvl in range(1, level + 1)]
                    )
                    nodes[value] = Node(value, parent=current_parent, section_id=section_id)
                current_parent = nodes[value]  # 更新父node

        # 添加文字内容为子节点
        text_value = item.get("text", "").strip()
        if text_value:
            section_id = " > ".join(
                [item.get(f"section_level_{lvl}", "").strip() for lvl in range(1, 8)]
            )
            Node(text_value, parent=current_parent, section_id=section_id)
    return root





# 構建樹
root_node = build_tree(content_ISO3)

# 打印樹狀結構
for pre, fill, node in RenderTree(root_node):
    # 如果有第二個 \n，移除其後的內容
    simplified_name = node.name.split("\n", 2)[0] if node.name.count("\n") >= 2 else node.name
    print(f"{pre}{simplified_name}")
```
